# Работете заедно, играјте паметно: Колективна интелегенција во екипите на играта League of Legends

Во овој труд се прави истражување за играта `League of Legends`. Оваа игра моментално е една од најпопуларните игри во светот со 115 милиони активни играчи во 2020 година. Главната причина за реализација на овој труд е да се утрврди ‚‚колективната интелегенција‘‘ за одредени тимови во играта `League of Legends`. Методите се извршуваа во лаборатории, со тестирање на поединци и на групи со цел да се измери колективната интелегенција. Тестирањето траеше еден час со што се извршуваа одредени задачи за интелегенција. Овие методи не се извршуваа надвор од лабораториите. Способноста на тимот непречено да работи заедно може да придонесе за нивниот перформанс во онлајн-тимски игри, како и во секоја тимска работа.

Во самата игра постојат дивизии според кои се класифицираат играчите според нивните вештини во играта. Дивизиите се поделени во 5 групи и тоа: **Bronze**, **Silver**, **Gold**, **Platinum** и **Diamond**. Колку подобри вештини толку појака дивизија односно, играчите со најслаби вештини припаѓаат во Bronze дивизија, додека пак играчите најдобри вештини спаѓаат во Diamond дивизија.

```python
import matplotlib.pyplot as plt
vals = [-0.2,0,0.1,0.4,0.7]
plt.plot(vals,color='red', linestyle='--', linewidth=3)
plt.grid(b=True, color='aqua', alpha=0.3, linestyle='-.', linewidth=2)
plt.show()
```

Да речеме дека броевите кои што се наоѓаат на х оската ги претставуваат дивизиите. Односно колку е помала вредноста по х оска толку е помала дивизијата. Тоа значи дека 0.0 претставува Bronze дивизија, 1.0 преставува Silver дивизија, 2.0 претставува Gold дивизија, 3.0 преставува Platinum дивизија и секако 4.0 претставува Diamond дивизија. 

Додека пак на у оската е претставена `колективната интелегенција`, односно ова значи дека колку е повисок резултатот толку подобра `колективна интелегенција` постои во тимовите.

Овој график ни покажува дека играчите кои што се наоѓаат во повисока дивизија имаат повисока `колективна интелегенција` и обратно, играчите кои што се наоѓаат во пониска дивизија имаат пониска `колективна интелегенција`.
